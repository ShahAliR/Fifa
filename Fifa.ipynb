```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows',None)

#data.sort_values(by='count')                       
#data.dtypes


data = pd.read_csv('fifa21_train.csv')
data.head()
data.tail()

```

```python
numerical = round(data.select_dtypes(np.number))
numerical
```

```python
categorical = data.select_dtypes(object)
categorical
```

```python
# Describing data, sorting by descending order
data.describe().T.sort_values(by='count')
data.isna().sum().sort_values(ascending=False)
```

```python
# Renaming columns: all lower, space replaced by underscore (_)
cols = []
for i in range(len(data.columns)):
    cols.append(data.columns[i].lower().replace(' ', '_'))
data.columns = cols
data.columns
```

```python
# Removing column loan_date_end: too many NaN values
data = data.drop(['loan_date_end'], axis=1)
```

```python
data = data.drop_duplicates()
data.shape # Only removed a few lines.
```
