# 1. Importing dataset

```python
import pandas as pd
import numpy as np
#pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows',None)

#data.sort_values(by='count')                       
#data.dtypes


data = pd.read_csv('fifa21_train.csv')
data.head()
data.tail()

```

# 2. Checking for types of values: numerical and categorical

```python
data.dtypes
```

```python
data.describe().T
```

# 3. Renaming columns: standardizing names

```python
# Renaming columns: all lower, space replaced by underscore (_)
cols = []
for i in range(len(data.columns)):
    cols.append(data.columns[i].lower().replace(' ', '_'))
data.columns = cols
data.columns
```

# 4. Analysing data: removing columns, dealing with NaN values

```python
# Describing data, sorting by descending order
data.describe().T.sort_values(by='count')
data.isna().sum().sort_values(ascending=False)
```

```python
# Checking for missing NaN values (percentage)
round(data.isna().sum()/len(data),4)*100
```

```python
# Removing column loan_date_end: too many NaN values
data = data.drop(['loan_date_end'], axis=1)
data.shape
```

```python
# Removing all NaN rows; we still have enough data points to work with. From 11.701 rows to 11.211 rows.
data = data.dropna()
data.shape # Only removed a few lines.
```

```python
# Checking if NaN values are still remaining
data.isna().sum()
```

# 5. Splitting numerical and categorical

```python
numerical = data.select_dtypes(np.number)
numerical.head()
```

```python
# From numerical: removing columns that we feel will not have an impact in our study: 'id'
numerical = numerical.drop(['id'], axis=1)
numerical.head()
```

```python
categorical = data.select_dtypes(object)
categorical.head()
```

```python
# From categorical: removing columns that we feel will not have an impact in our study: 'name', 'nationality', 
# 'club', 'team_&_contract', 'joined', 'contract'
# Also removing 'position' because we have very similar information in column 'bp'

categorical = categorical.drop(['position', 'name', 'nationality', 'club', 'team_&_contract', 'joined', 'contract'], axis=1)
categorical.head()
```

```python
categorical['bp'].unique()
```

# 6. Checking for outliers

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Using 'ova' (Overall Rating) to see if our target contains any outliers that we should be aware of.
# As we can see in the graph there are a few outliers that we assume are player that are either very good or very bad.
# We will not be removing these outliers.
sns.boxplot(x=data['ova'])
plt.show()
```

```python
fig, ax=plt.subplots(figsize=(40,32)) # Used to set the size of the box. (x axis,y axis)

correlations_matrix = round(numerical.corr(),3)
sns.heatmap(correlations_matrix, annot=True)
plt.show()
```
